```python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}
```

```python
def find_doc(doc_number):        # функция возвращает индекс документа с номером doc_number либо -1 если документ не найден'
    for document in documents:
        if document['number'] == doc_number:
            return documents.index(document)
    return -1

def find_shelf(doc_number):      # функция возвращает ключ полки документа с номером doc_number либо '' если документ не найден
    for shelf, docs in directories.items():
        for doc in docs:
            if doc == doc_number:
                return shelf
    return ''

def doc_list():                  # функция возвращает список документов
    result = []
    for doc in documents:
        result.append(f"№: {doc['number']}, тип: {doc['type']}, владелец: {doc['name']}, полка хранения: {find_shelf(doc['number'])}")
    return '\n'.join(result)

def shelf_list():                # функция возвращает текущий перечень полок
    return f"Текущий перечень полок: {str(sorted([int(key) for key in directories.keys()])).strip('[]')}"


while True:
    cmd = input('Введите команду:\n').lower()
    if cmd == 'p':
        doc_index = find_doc(input('Введите номер документа:\n'))
        if doc_index != -1:
            output = f"Владелец документа: {documents[doc_index]['name']}"
        else:
            output = 'Документ не найден в базе'
    elif cmd == 's':
        shelf = find_shelf(input('Введите номер документа:\n'))
        if shelf != '':
            output = f'Документ хранится на полке: {shelf}'
        else:
            output = 'Документ не найден в базе'
    elif cmd == 'l':
        output = doc_list()
    elif cmd == 'as':
        shelf = input('Введите номер полки:\n')
        if shelf in directories:
            output = f'Такая полка уже существует. {shelf_list()}'
        else:
            directories[shelf] = []
            output = f'Полка добавлена. {shelf_list()}'
    elif cmd == 'ds':    
        shelf = input('Введите номер полки:\n')
        if shelf in directories:
            if directories[shelf] == []:
                del(directories[shelf])
                output = f'Полка удалена. {shelf_list()}'
            else:
                output = f'На полке есть документы, удалите их перед удалением полки. {shelf_list()}'
        else:
            output = f'Такой полки не существует. {shelf_list()}'
    elif cmd == 'ad':
        doc = {'number' : input('Введите номер документа:\n'),
               'type' : input('Введите тип документа:\n'),
               'name' : input('Введите владельца документа:\n')
              }
        shelf = input('Введите номер полки:\n')
        if shelf in directories:
            documents.append(doc)
            directories[shelf].append(doc['number'])
            output = f'Документ добавлен. Текущий список документов:\n{doc_list()}'
        else:
            output = 'Такой полки не существует. Добавьте полку командой as.'
    elif cmd == 'd':
        num_doc = input('Введите номер документа:\n')
        doc_index = find_doc(num_doc)
        doc_shelf = find_shelf(num_doc)
        if doc_index != -1:
            documents.pop(doc_index)
            directories[doc_shelf].remove(num_doc)
            output = f'Документ удален. Текущий список документов:\n{doc_list()}'
        else: 
            output = f'Документ не найден в базе. Текущий список документов:\n{doc_list()}'
    elif cmd == 'm':
        num_doc = input('Введите номер документа:\n')
        doc_index = find_doc(num_doc)
        old_shelf = find_shelf(num_doc)
        new_shelf = input('Введите номер полки:\n')
        if doc_index != -1:
            if new_shelf in directories:
                directories[old_shelf].remove(num_doc)
                directories[new_shelf].append(num_doc)
                output = f'Документ перемещен. Текущий список документов:\n{doc_list()}'
            else:
                output = f'Такой полки не существует. {shelf_list()}'
        else: 
            output = f'Документ не найден в базе. Текущий список документов:\n{doc_list()}'
    elif cmd == 'q':
        break
    else:
        output = ('p   - узнать владельца документа по номеру\n' 
                  's   - узнать полку документа по номеру\n' 
                  'l   - полная информация по всем документам\n'
                  'as  - добавить новую полку\n'
                  'ds  - удалить полку\n'
                  'ad  - добавить новый документ\n'
                  'd   - удалить документ\n'
                  'm   - переместить документ с полки на полку'
                  'q   - выход')
    print(output)
```
