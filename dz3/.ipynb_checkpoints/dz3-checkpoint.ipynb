```python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}

```

```python
def find_doc(doc_number):
    for document in documents:
        if document['number'] == doc_number:
            return document['name']
    return 'Документ не найден в базе'

def find_shelf(doc_number):
    for shelf, docs in directories.items():
        for doc in docs:
            if doc == doc_number:
                return shelf
    return 'Документ не найден в базе'

def doc_list():
    result = []
    for doc in documents:
        result.append(f'№: {doc['number']}, тип: {}, владелец: {}, полка хранения: {}')
    return result
```

```python
def input_command():
    input('Введите команду:', cmd)
    if cmd in ['n','s','l']:
        return cmd, true
    elif cmd == 'q':
        return cmd, false
    else:
        return '', true


while input_command()[1]:
    if command == 'n':
        doc_number = input('Введите номер документа:')
        for document in documents:
            if document['number'] == doc_number:
                return print(f'Владелец документа {document['name']}')
        return print('Документ не найден в базе')
    elif command == 's':
        doc_number = input('Введите номер документа:')
        for shelf, docs in directories.items():
            for doc in docs:
                if doc == doc_number:
                    return print(f'Документ хранится на полке: {shelf}')
        return print('Документ не найден в базе')
    elif command == 'l':
        result = []
        for doc in documents:
            result.append(f'№: {doc['number']}, тип: {}, владелец: {}, полка хранения: {}')
        return print(result)
    else:
        print('Такой команды нет')
```
