```python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}
```

```python
def find_doc_index(doc_number):        # функция возвращает индекс документа с номером doc_number либо -1 если документ не найден'
    for document in documents:
        if document['number'] == doc_number:
            return documents.index(document)
    return -1

def find_shelf_key(doc_number):      # функция возвращает ключ полки документа с номером doc_number либо '' если документ не найден
    for shelf, docs in directories.items():
        for doc in docs:
            if doc == doc_number:
                return shelf
    return ''

def doc_list():                  # функция возвращает список документов
    result = []
    for doc in documents:
        result.append(f"№: {doc['number']}, тип: {doc['type']}, владелец: {doc['name']}, полка хранения: {find_shelf_key(doc['number'])}")
    return '\n'.join(result)

def shelf_list():                # функция возвращает текущий перечень полок
    return f"Текущий перечень полок: {str(sorted([int(key) for key in directories.keys()])).strip('[]')}"

def find_name_by_doc(doc_number):     # находит владельца по номеру документа
    doc_index = find_doc_index(doc_number)
    if doc_index != -1:
        return f"Владелец документа: {documents[doc_index]['name']}"
    else:
        return 'Документ не найден в базе'

def find_shelf_by_doc(doc_number):    # находит полку по номеру дркумента
    shelf = find_shelf_key(doc_number)
    if shelf != '':
        return f'Документ хранится на полке: {shelf}'
    else:
        return 'Документ не найден в базе'

def add_shelf(new_shelf):              # добавляет новую полку
    if new_shelf in directories:
        return f'Такая полка уже существует. {shelf_list()}'
    else:
        directories[new_shelf] = []
        return f'Полка добавлена. {shelf_list()}'

def del_shelf(shelf):                 # удаляет существующую полку
    if shelf in directories:
        if directories[shelf] == []:
            del(directories[shelf])
            return f'Полка удалена. {shelf_list()}'
        else:
            return f'На полке есть документы, удалите их перед удалением полки. {shelf_list()}'
    else:
        return f'Такой полки не существует. {shelf_list()}'    

def add_doc(num_doc, type_doc, name, shelf):  # добавляет новый документ на существующую полку
    doc = {'number' : num_doc, 'type' : type_doc, 'name' : name}
    if shelf in directories:
        documents.append(doc)
        directories[shelf].append(doc['number'])
        return f'Документ добавлен. Текущий список документов:\n{doc_list()}'
    else:
        return 'Такой полки не существует. Добавьте полку командой as.'    

def del_doc(num_doc):                      # удаляет документ
    doc_index = find_doc_index(num_doc)
    doc_shelf = find_shelf_key(num_doc)
    if doc_index != -1:
        documents.pop(doc_index)
        directories[doc_shelf].remove(num_doc)
        return f'Документ удален. Текущий список документов:\n{doc_list()}'
    else: 
        return f'Документ не найден в базе. Текущий список документов:\n{doc_list()}'

def move_doc(num_doc, new_shelf):          # перемещает документ на существующую полку
    doc_index = find_doc_index(num_doc)
    old_shelf = find_shelf_key(num_doc)
    if doc_index != -1:
        if new_shelf in directories:
            directories[old_shelf].remove(num_doc)
            directories[new_shelf].append(num_doc)
            return f'Документ перемещен. Текущий список документов:\n{doc_list()}'
        else:
            return f'Такой полки не существует. {shelf_list()}'
    else: 
        return f'Документ не найден в базе. Текущий список документов:\n{doc_list()}'

def info():                            # подсказка
    return ('p   - узнать владельца документа по номеру\n' 
            's   - узнать полку документа по номеру\n' 
            'l   - полная информация по всем документам\n'
            'as  - добавить новую полку\n'
            'ds  - удалить полку\n'
            'ad  - добавить новый документ\n'
            'd   - удалить документ\n'
            'm   - переместить документ с полки на полку'
            'q   - выход')
    
def main():  
    while True:
        cmd = input('Введите команду:\n').lower()
        if cmd == 'p':
            output = find_name_by_doc(input('Введите номер документа:\n'))
        elif cmd == 's':
            output = find_shelf_by_doc(input('Введите номер документа:\n'))
        elif cmd == 'l':
            output = doc_list()
        elif cmd == 'as':
            output = add_shelf(input('Введите номер полки:\n'))
        elif cmd == 'ds':
            output = del_shelf(input('Введите номер полки:\n'))
        elif cmd == 'ad':
            output = add_doc(input('Введите номер документа:\n'),
                             input('Введите тип документа:\n'),
                             input('Введите владельца документа:\n'),
                             input('Введите номер полки:\n')
                            )
        elif cmd == 'd':
            output = del_doc(input('Введите номер документа:\n'))
        elif cmd == 'm':
            output = move_doc(input('Введите номер документа:\n'), input('Введите номер полки:\n'))
        elif cmd == 'q':
            break
        else:
            output = info()
        print(output)
main()
```
