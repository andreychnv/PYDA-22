```python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}

```

```python
def find_name(doc_number):                  # функция возвращает имя владельца документа doc_number или 'Документ не найден'
    for document in documents:
        if document['number'] == doc_number:
            return f"Владелец документа: {document['name']}"
    return 'Документ не найден в базе'

def find_shelf(doc_number):                 # функция возвращает полку документа doc_number или 'Документ не найден'
    for shelf, docs in directories.items():
        for doc in docs:
            if doc == doc_number:
                return f'полка хранения: {shelf}'
    return 'Документ не найден в базе'

def doc_list():                             # функция возвращает список документов
    result = []
    for doc in documents:
        result.append(f"№: {doc['number']}, тип: {doc['type']}, владелец: {doc['name']}, {find_shelf(doc['number'])}")
    return '\n'.join(result)

def shelf_list():                           # функция возвращает текущий перечень полок
    return f"Текущий перечень полок: {str(sorted([int(key) for key in directories.keys()])).strip('[]')}"

def add_shelf(shelf):                       # функция добавляет полку или сообщает, что она уже есть
    if shelf in directories:
        return f'Такая полка уже существует. {shelf_list()}'
    else:
        directories[shelf] = []
        return f'Полка добавлена. {shelf_list()}'

def del_shelf(shelf):                       # функция удаляет существующую пустую полку или сообщает, что она не пустая
    if len(directories[shelf]) == 0:
        del(directories[shelf])
        return f'Полка удалена. {shelf_list()}'
    else:
        return f'На полке есть документы, удалите их перед удалением полки. {shelf_list()}'

def add_doc(doc_number, doc_type, doc_name, doc_shelf):
    if doc_shelf in directories:
        directories[doc_shelf].append(doc_number)
        documents.append({'type' : doc_type, 'number' : doc_number, 'name' : doc_name})
        return f'Документ добавлен. Текущий список документов:\n{doc_list()}'
    else:
        return 'Такой полки не существует. Добавьте полку командой as.'

def del_doc(doc_number):
    for document in documents:
        if document['number'] == doc_number:
            documents.remove(document)
            for shelf, docs in directories.items():
                for doc in docs:
                    if doc == doc_number:
                        docs.remove(doc)
                        return f'Документ удален. Текущий список документов:\n{doc_list()}'
    return f'Документ не найден в базе. Текущий список документов:\n{doc_list()}'

def mov_doc(doc_number, doc_shelf):
    

    
    
    
    
while True:
    cmd = input('Введите команду:\n').lower()
    if cmd == 'p':
        output = find_name(input('Введите номер документа:\n'))
    elif cmd == 's':
        output = find_shelf(input('Введите номер документа:\n')).capitalize()
    elif cmd == 'l':
        output = doc_list()
    elif cmd == 'as':
        output = add_shelf(input('Введите номер полки:\n'))
    elif cmd == 'ds':    
        shelf = input('Введите номер полки:\n')
        if shelf in directories:
            output = del_shelf(shelf)
        else:
            output = f'Такой полки не существует. {shelf_list()}'
    elif cmd == 'ad':
        output = add_doc(input('Введите номер документа:\n'),
                         input('Введите тип документа:\n'),
                         input('Введите владельца документа:\n'),
                         input('Введите номер полки:\n'))
    elif cmd == 'd':                
        output = del_doc(input('Введите номер документа:\n'))
    elif cmd == 'm':
        
        output = mov_doc(input('Введите номер документа\n', input('Введите номер полки:\n')))
    elif cmd == 'q':
        break
    else:
        output = ('p   - узнать владельца документа по номеру\n' 
                  's   - узнать полку документа по номеру\n' 
                  'l   - полная информация по всем документам\n'
                  'as  - добавить новую полку\n'
                  'ds  - удалить полку\n'
                  'ad  - добавить новый документ\n'
                  'd   - удалить документ\n'
                  'm   - переместить документ с полки на полку'
                  'q   - выход')
    print(output)
```
