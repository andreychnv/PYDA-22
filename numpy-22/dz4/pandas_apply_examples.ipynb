# Создание столбцов на ходу
- Лена - разработчик
- Ваня - дизайнер
- Настя - редактор

```python
import pandas as pd
```

```python
staff = {
    'Лена': 'разработчик',
    'Ваня': 'дизайнер',
    'Настя': 'редактор',
}
```

```python
def departments(row):
    dict_ = row['dicts']
    
    if 'Лена' in dict_:
        row['разработчик'] = dict_['Лена']
    
    if 'Настя' in dict_:
        row['редактор'] = dict_['Настя']
    
    return row
```

```python
def departments(row):
    dict_ = row['dicts']
    
    for name, tasks in dict_.items():
        if name in staff:
            row[staff[name]] = dict_[name]
            
    return row
```
