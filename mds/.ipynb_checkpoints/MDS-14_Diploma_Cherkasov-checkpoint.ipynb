## Задание 1

Дана матрица рейтингов фильмов (строки - пользователи (10 человек), столбцы - фильмы (15 фильмов))

```python
raitings = np.array([[ 4,  4,  9,  4,  1,  6, 10,  7,  9,  6,  9,  2,  8,  6,  6],
                     [ 9,  2,  5, 10,  7,  8, 10,  5,  6,  2,  1,  6,  8,  9,  7],
                     [ 1,  6,  8,  8,  4,  9,  3,  8, 10,  5,  2,  6,  8,  1,  6],
                     [ 6,  1,  9,  7,  7,  9,  2,  3,  5,  1,  6,  6,  3,  2,  7],
                     [ 3,  7,  3,  5,  7,  9,  9,  6,  2,  9,  1,  2,  8, 10,  6],
                     [ 8,  3,  7,  3,  8,  6,  1,  8,  8,  6,  1,  9,  4, 10,  1],
                     [ 9,  8,  4,  8,  8, 10,  6,  1,  1,  2,  9,  5,  2,  7,  2],
                     [ 4,  1,  6,  4,  3, 10,  4,  4,  2,  8,  7,  9,  3,  8,  3],
                     [ 2,  7,  7,  6, 10,  6,  8,  9,  8,  6, 10,  1,  7, 10,  4],
                     [ 5, 10,  8,  8,  9,  7,  2,  9,  9, 10,  8,  8,  8,  6, 10]])
```

**a)** Необходимо найти наиболее похожего пользователя по косинусной метрике для каждого из (ответ: 10 пар вида (x, y), где y - наиболее похожий пользователь на пользователя x).

**b)** Предположим, вы захотели сжать признаковые описания пользователя с 15 фильмов, до, скажем, 5-ти признаков, как вы это сделаете?

```python
#a)
from sklearn.metrics import pairwise_distances

cos_sim = pairwise_distances(raitings, metric="cosine")  # матрица расстояний по cos-метрике
for i in range(len(cos_sim)):
    cos_sim[i][i] = 1                                    # максимизируем значения расстояний до самого себя
pair = []
for i in range(len(cos_sim)):
    pair.append([i, np.argmin(cos_sim[i])])              # для каждой строки ищем индекс минимального значения
print(pair)

```

```python
#b)
from sklearn.decomposition import TruncatedSVD
svd = TruncatedSVD(n_components=5)
svd.fit_transform(raitings)
```

## Задание 3

Дана функция с неизвестными параметрами a и b, вам также известны значения функции fx в точках x. Найдите оптимальные параметры a, b, минимизирующие абсолютную ошибку **error**.

```python
def f(x, a, b):
    return np.e**a * np.sin(b*x) + x
```

```python
x = np.array([0.        , 0.26315789, 0.52631579, 0.78947368, 1.05263158,
              1.31578947, 1.57894737, 1.84210526, 2.10526316, 2.36842105,
              2.63157895, 2.89473684, 3.15789474, 3.42105263, 3.68421053,
              3.94736842, 4.21052632, 4.47368421, 4.73684211, 5.        ])
```

```python
fx = np.array([  0.        , -12.01819092, -18.90968634, -17.68786571,
                -8.7529108 ,   4.27524517,  16.06801336,  21.81250213,
                19.22059845,   9.48411207,  -3.22273056, -13.48576488,
               -16.91096359, -11.95866834,  -0.58630088,  12.56873816,
                22.12489421,  24.20292139,  18.04522521,   6.33211092])
```

```python
plt.figure(figsize=(20,10))
plt.plot(x, fx)

plt.show()
```

```python
def error(params):
    return np.sum(np.abs(fx - f(x, params[0], params[1])))
```

```python
from scipy.optimize import minimize
res = minimize(error, [0, 0], method='COBYLA', tol=1e-6)
params = res.x
print(res.x)
plt.figure(figsize=(20,10))

plt.plot(x, f(x, res.x[0], res.x[1]), linestyle = '--', color = 'red', label = f'f(x), a={res.x[0] :.3f}, b={res.x[1] :.3f} (optimized)')
plt.plot(x, f(x, 2.5, -3), label = 'f(x), a=2.5, b=-3')
plt.plot(x, f(x, 3.5, -2), label = 'f(x), a=3.5, b=-2')
plt.plot(x, fx, linestyle = '-.', color = 'lightgray', label = 'fx')
plt.legend()

```

## Задание 5

Найдите собственные значения и собственные вектора матрицы (на python и на бумаге):

```python
m = np.array([[1, 4],
              [1, 1]])
print(m)

# |1-l   4 |
# |        | = 0    =>  l**2 - 2 * l - 3 = 0    l = [3, -1]
# | 1   1-l|
#
# l[0] = 3                     l[1] = -1
# |-2x + 4y = 0                | 2x + 4y = 0
# |  x - 2y = 0                |  x + 2y = 0
# x, y = [2, 1]                x, y = [2, -1]
#
# v[0] = [2, 1]                v[1] = [2, -1]
#

w, v = np.linalg.eig(m)
print(w)
print(v)
```
