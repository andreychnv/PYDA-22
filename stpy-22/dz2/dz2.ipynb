```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# соберем данные за все годы
df2015 = pd.read_csv('2015.csv')
df2016 = pd.read_csv('2016.csv')
df2017 = pd.read_csv('2017.csv')
df2018 = pd.read_csv('2018.csv')
df2019 = pd.read_csv('2019.csv')
```

```python
# приведем датафреймы к единообразию и соединим
df2015.columns = ['Country','Region','Rank','Happiness Score','2','GDP per capita','Social support','Healthy life expectancy','Freedom','Trust','Generosity','Other']
df2016.columns = ['Country','Region','Rank','Happiness Score','2','3','GDP per capita','Social support','Healthy life expectancy','Freedom','Trust','Generosity','Other']
df2017.columns = ['Country','Rank','Happiness Score','1','2','GDP per capita','Social support','Healthy life expectancy','Freedom','Generosity','Trust','Other']
df2018.columns = ['Rank','Country','Happiness Score','GDP per capita','Social support','Healthy life expectancy','Freedom','Generosity','Trust']
df2019.columns = ['Rank','Country','Happiness Score','GDP per capita','Social support','Healthy life expectancy','Freedom','Generosity','Trust']
df2015['Year'] = 2015
df2016['Year'] = 2016
df2017['Year'] = 2017
df2018['Year'] = 2018
df2019['Year'] = 2019
df = pd.concat([df2015, df2016, df2017, df2018, df2019])
```

```python
# дозаполним недостающие регионы на основе данных из 2015, 2016 годов и удалим ненужные колонки
dfreg = df[['Country','Region']].loc[~df['Region'].isnull()].groupby(['Country']).max()
df.drop(['Region','1','2','3'], axis=1, inplace=True) 
df = df.merge(dfreg, how='inner', left_on='Country', right_index=True)
df['Other'] = (df['Happiness Score'] 
               - df['GDP per capita'] 
               - df['Social support']
               - df['Healthy life expectancy'] 
               - df['Freedom'] 
               - df['Trust'] 
               - df['Generosity']
              )
```

```python
# список годов
Years = list(df['Year'].groupby(df['Year']).count().index)
# словари лучших рейтингов стран, лучших и худших регионов по годам
CountryTop5 = dict()
for year in Years:
    CountryTop5[year] = list(df.loc[df['Year']==year].sort_values('Happiness Score', ascending=False).head(5)['Country'])
RegionTop5 = dict()
for year in Years:
    RegionTop5[year] = list(df.loc[df['Year']==year].groupby(['Region']).mean().sort_values('Happiness Score', ascending=False).head(5).index)

RegionBottom5 = dict()
for year in Years:
    RegionBottom5[year] = list(df.loc[df['Year']==year].groupby(['Region']).mean().sort_values('Happiness Score', ascending=True).head(5).index)

    
indicators = ['GDP per capita', 'Social support', 'Healthy life expectancy', 'Freedom', 'Trust', 'Generosity']
```

```python
year = 2019
(pd.concat([df.loc[(df['Region'].isin(RegionTop5[year][:3])) & (df['Year']==year),
                   ['Region'] + ['Happiness Score'] + indicators].groupby(['Region']).mean(),
            df.loc[(df['Region'].isin(RegionBottom5[year][:3])) & (df['Year']==year),
                   ['Region'] + ['Happiness Score'] + indicators].groupby(['Region']).mean()
           ]
          )
 .sort_values(['Happiness Score'], ascending=False)
 .drop(['Happiness Score'], axis=1)
 .plot(kind='bar',
       ylabel='Happiness Score',
       title='Happiness Scores by TOP-5 Regions in 2019',
       grid=True,
       figsize=(10,5))
)
```

#### на визуализации можно предположить, 
#### что есть некоторая зависимость между показателями: 
#### продолжительность жизни, ВВП, меры социальной поддержки и индексом счастья

```python
# выбрать год и страну с максимальным общим показателем
year = 2019
country = CountryTop5[year][0]

(df
 .loc[df['Country'].isin(CountryTop5[year]), ['Country', 'Year', 'Happiness Score']]
 .pivot(index='Year', columns='Country', values='Happiness Score')
 .plot(title='Dynamics of the happiness indicator by country',
       ylabel='Happiness Score',
       grid=True,
       xticks=Years
 )
)
```
