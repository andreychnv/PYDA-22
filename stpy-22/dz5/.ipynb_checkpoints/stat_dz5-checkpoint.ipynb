#### Классификация

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.metrics import accuracy_score
import pandas as pd
import matplotlib.pyplot as plt
```

```python
iris  = load_iris()
print(iris.DESCR)
```

```python
iris.feature_names
```

```python
data = pd.DataFrame(iris.data[:,:2], columns=iris.feature_names[:2])
data_class = iris.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, data_class, test_size=0.25)
```

```python
lda = LinearDiscriminantAnalysis()
```

```python
lda.fit(X_train, y_train)
```

```python
classes = lda.predict(X_test)
```

```python
result = pd.DataFrame([y_test, classes]).T
```

```python
result
```

ошибок много, узнаем точность

```python
accuracy_score(y_test, classes)
```

сравним визуально

```python
fig, axes = plt.subplots(1,2)
axes[0].set_title('"LDA"-model')
axes[0].scatter(X_test['sepal length (cm)'], X_test['sepal width (cm)'], c=classes)
axes[0].scatter(lda.means_[:, 0], lda.means_[:, 1], c='r', s=150, marker='*')
axes[1].set_title('original')
axes[1].scatter(X_test['sepal length (cm)'], X_test['sepal width (cm)'], c=y_test)
fig.set_figwidth(12)
plt.show()
```

видим, что модель дала много ошибок при определении классов.
Можно предположить, что в построенной модели для отнесения к классу 0 (слева вверху) 
выбранные признаки являются довольно существенными,
а для отнесения к 1 и 2 классам необходимо учитывать дополнительные признаки

#### Кластеризация

```python
from sklearn.preprocessing import StandardScaler
import numpy as np
from sklearn.cluster import KMeans
scaler = StandardScaler()
```

переводим датафрейм в массив

```python
X = np.array(data)
```

масштабируем

```python
X_scaled = scaler.fit_transform(X)
```

```python
plt.scatter(X_scaled[:, 0], X_scaled[:, 1],cmap='autumn')
```

вычисляем массив показателей инерции

```python
k_inertia = []
ks = range(1, 20)

for k in ks:
    clf_kmeans = KMeans(n_clusters=k)
    clusters_kmeans = clf_kmeans.fit_predict(X_scaled)
    k_inertia.append(clf_kmeans.inertia_)
```

строим график показателя инерции от числа кластеров

```python
plt.plot(ks, k_inertia)
plt.plot(ks, k_inertia ,'ro')
```

наибольший "излом" в точке 3 -> 3 кластера

строим модель для 3 кластеров

```python
kmeans = KMeans(n_clusters=3)
clusters = kmeans.fit_predict(X_scaled)
clusters
```

рисуем график

```python
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=clusters)
```

```python
fig, axes = plt.subplots(1,2)
axes[0].set_title('"K-Means"-model')
axes[0].scatter(X[:,0], X[:,1], c=clusters)
axes[1].set_title('original')
axes[1].scatter(X[:,0], X[:,1], c=data_class)
fig.set_figwidth(12)
plt.show()
```

ситуация, схожая с LDA-моделированием: класс "0" определяется по выбранным показателям гораздо точнее, чем "1" и "2"

сравним все 3 варианта

```python
fig, axes = plt.subplots(1,3)
axes[0].set_title('original')
axes[0].scatter(X[:,0], X[:,1], c=data_class)
axes[1].set_title('"LDA"-model')
axes[1].scatter(X[:,0], X[:,1],  
                c=lda.predict(X)
               )
axes[2].set_title('"K-Means"-model')
axes[2].scatter(X[:,0], X[:,1], c=clusters)
fig.set_figwidth(15)
plt.show()
```

График LDA-модели немногим более точен за счет того, что:
1. в моделировании использовались не все признаки, на основе которых классифицируются объекты
2. 75% верных результатов LDA-модель получила в качестве данных для обучения
